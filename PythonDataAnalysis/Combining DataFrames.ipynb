```python
import pandas as pd
surveys_df = pd.read_csv("data/surveys.csv")
```

```python
surveys_head = surveys_df.head(10)

surveys_tail = surveys_df.tail(10).reset_index(drop=True)   # reset_index reset the index of tail data
```

```python
vertical_stack = pd.concat([surveys_head,surveys_tail, axis=0]).reset_index(drop=True)
vertical_stack                                           # axis =0 is passed in list instead 
                                                        # as parameter, drop = True if don't want
                                                        # extra cols
```
