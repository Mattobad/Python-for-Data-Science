## Matplotlib
### > Low-level graphics library
### > Gives large amount of control over the plots
### > Often times might be tedious
### > Similar to Matlab's graphical plotting library

```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline  
# this should be included in Jupyter otherwise graph will not show up
```

```python
x = np.linspace(0,np.pi*4,num=200)
```

```python
y = np.sin(x)
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y)
plt.show()
```

```python
z = np.cos(x)

plt.plot(x,y,'r',label='sin(x)')
plt.plot(x,z,'b',label='cos(x)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.show()
```

```python
z = np.cos(x)

plt.plot(x,y,'r',label='sin(x)')
plt.plot(x,z,'b',label='cos(x)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.xlim([0,6])
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(x,z,'b')

plt.tight_layout()  # helps to overcome the overlapping between the graphs
plt.show()
```

```python
fig = plt.figure()  # working as oop

axes = fig.add_axes([0.1,0.1,0.8,0.8])  # coordinates (x,y,height,width)

axes.plot(x,y,label='sin(x)')
axes.legend()

axes.set_xlabel('x')
axes.set_ylabel('sin(x)')
axes.set_title('Graph')

plt.show()
```

### Histogram, scatter, 2d rectangular, and hexagonal histograms using Matplotlib

```python
import pandas as pd
import numpy as np
```

```python
col_names =['mpg','cylinders','dis','hp','weight','acc','year','origin','name']
```

```python
df = pd.read_csv(r'C:\Users\s6040865\Downloads\auto-mpg.data',sep='\s+',names=col_names)
df.to_csv(r'C:\Users\s6040865\Downloads\auto-clean.csv')  # writing to csv file
df.head()
```

```python
plt.hist(df['mpg'])
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.hist(df['mpg'],bins='auto')  #plt.hist(df['mpg'],bins=20)
plt.xlabel('mpg')
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.scatter(x=df['weight'],y=df['mpg'], alpha=0.5)

plt.ylabel('mpg')
plt.xlabel('weight')
plt.show()                    # shows negative correlation between weight & mpg when weight increases mpg decreases
```

```python
plt.figure(figsize=(10,6))
plt.hist2d(x=df['weight'],y=df['mpg'])   #see the difference when putting parameter alpha = 0.5

plt.ylabel('mpg')
plt.xlabel('weight')
plt.colorbar()              #its better use it when you have lot of points overlapped together
plt.show()    
```

```python
plt.figure(figsize=(10,6))
plt.hexbin(x=df['weight'],y=df['mpg'], gridsize=(35,15), cmap='coolwarm')

plt.ylabel('mpg')
plt.xlabel('weight')
plt.colorbar()
plt.show()    
```

## Produce histogram,scatter, bar, and box plots using Seaborn
### **SeaBorn**
  ### . Based on Matplotlib
  ### . High-level plotting
  ### . Easier to work with compared to Matplotlib

```python
import seaborn as sns
import warnings

warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
df = pd.read_csv(r'C:\Users\s6040865\Downloads\auto-clean.csv')
df.head()
```

```python
sns.distplot(df['mpg'])
```

```python
sns.distplot(df['mpg'], kde=False, bins=20)
plt.show()
```

```python
plt.figure(figsize=[10,6])

sns.distplot(df['mpg'], kde=False, bins=20)
plt.show()
```

```python
sns.lmplot('weight','mpg',data=df, aspect=2)
```

```python
sns.lmplot('weight','mpg',data=df,fit_reg=False,aspect=2, size=4)
```

```python
sns.lmplot('weight','mpg',data=df,fit_reg=False,aspect=2, size=4, hue='cylinders')
```

```python
sns.lmplot('weight','mpg',data=df,fit_reg=False,aspect=1, size=5, hue='cylinders', col='origin')
```

```python
sns.kdeplot(df['weight'],df['mpg'], shade=True)  #kernel estimation plot
```

```python
sns.jointplot('weight','mpg',data=df, alpha =0.4)
```

```python
g = sns.PairGrid(df,diag_sharey=False)
g.map_lower(sns.kdeplot)
g.map_upper(plt.scatter)
g.map_diag(sns.kdeplot)
```

```python
sns.countplot(y='cylinders',data=df)
```

```python
sns.boxplot(x='cylinders',y='mpg',data=df)
```

```python
sns.boxplot(x='origin',y='mpg',data=df, hue='cylinders')
```

## Graphs with Pandas
### > Directly produce graphs by using pandas plotting methods
### > Quick and easy (sometimes dirty) solution

```python
df = pd.read_csv(r'C:\Users\s6040865\Downloads\auto-clean.csv')
df.head() 
```

```python
df['weight'].plot(kind='hist',bins=20)
```

```python
df.head()
```

```python
g =df.groupby('cylinders')[['mpg','cylinders','weight']]   #??? what groupby cylinders does here
g.head()
```

```python
df.groupby('cylinders')['mpg'].plot(kind='hist',alpha=0.6,legend=True)
```

```python
df.plot(kind='scatter',x='weight',y='mpg')
plt.show()
```

```python
df.boxplot(column='mpg',by='cylinders',grid=False)
plt.show()
```

```python
df.pivot_table(index='cylinders', aggfunc='mean')
```

```python
dfp =df.pivot_table(index='cylinders', aggfunc='mean')
```

```python
dfp.plot.line(y='acc')
```

```python
for col in dfp.columns:
    dfp.plot.line(y=col)
```

```python
df.groupby('cylinders').count()['mpg'].plot(kind='bar')
plt.show()
```
